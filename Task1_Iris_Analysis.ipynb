# Iris Dataset Analysis

## Introduction
The objective of this task is to explore and visualize the Iris dataset using Python. 
The dataset contains measurements of iris flowers including sepal length, sepal width, petal length, and petal width. 
Using visualization techniques we analyze relationships between variables and understand the dataset structure.

## Dataset Description

The Iris dataset contains 150 observations and 5 variables:

- Sepal Length
- Sepal Width
- Petal Length
- Petal Width
- Species

The species include:
- Setosa
- Versicolor
- Virginica

## Conclusion

Through exploratory data analysis we observed the relationships between different flower measurements.

Scatter plots showed that petal length and petal width clearly separate species.
Histograms displayed the distribution of each feature.
Box plots helped identify the spread and possible outliers.

This analysis demonstrates how visualization helps understand patterns in data.
